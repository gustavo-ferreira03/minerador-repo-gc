```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
repos_data = [
    ["Veloren", 5400, 5400, 15746, "Jogo"],
    ["Cataclysm: DDA", 9700, 4100, 113715, "Jogo"],
    ["Shattered Pixel Dungeon", 7200, 1000, 7328, "Jogo"],
    ["Mindustry", 21300, 2800, 17752, "Jogo"],
    ["Dungeon Crawl: Stone Soup", 2200, 1200, 69486, "Jogo"],
    ["Warzone 2100", 3100, 515, 20106, "Jogo"],
    ["Unciv", 7800, 1500, 10451, "Jogo"],
    ["OpenTTD", 6000, 834, 29448, "Jogo"],
    ["Endless Sky", 5400, 986, 8584, "Jogo"],
    ["The Battle for Wesnoth", 5300, 985, 83429, "Jogo"],
    ["Lichess", 14800, 2200, 60678, "Web"],
    ["Flutter Docs", 2700, 3200, 6742, "Web"],
    ["Monkeytype", 14400, 2200, 10883, "Web"],
    ["React Docs", 10800, 7400, 6533, "Web"],
    ["Ethereum.org", 4900, 4700, 34500, "Web"],
    ["Kubernetes Docs", 4400, 14100, 49543, "Web"],
    ["The Odin Project", 3500, 2100, 6008, "Web"],
    ["Tailwind Docs", 3200, 1700, 4837, "Web"],
    ["Bitwarden", 8600, 1100, 15275, "Web"],
    ["Privacy Guides", 2500, 196, 3470, "Web"]
]
```

```python
repos_array = np.array(repos_data)

games_data = repos_array[repos_array[:, 4] == "Jogo"]
web_data = repos_array[repos_array[:, 4] == "Web"]

games_names = games_data[:, 0]
games_stars = games_data[:, 1].astype(int)
games_forks = games_data[:, 2].astype(int)
games_commits = games_data[:, 3].astype(int)

web_names = web_data[:, 0]
web_stars = web_data[:, 1].astype(int)
web_forks = web_data[:, 2].astype(int)
web_commits = web_data[:, 3].astype(int)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(games_stars, games_commits, marker='x', color='blue', label='Jogo')
plt.scatter(web_stars, web_commits, marker='x', color='red', label='Web')
plt.title('Commits vs Estrelas')
plt.xlabel('Estrelas')
plt.ylabel('Commits')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(games_forks, games_commits, marker='x', color='blue', label='Jogo')
plt.scatter(web_forks, web_commits, marker='x', color='red', label='Web')
plt.title('Commits vs Forks')
plt.xlabel('Forks')
plt.ylabel('Commits')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
width = 0.4
x = np.arange(len(games_names))

plt.bar(x - width/2, games_stars, width, label='Estrelas (Jogo)', color='blue')
plt.bar(x + width/2, games_forks, width, label='Forks (Jogo)', color='lightblue')

plt.xticks(x, games_names, rotation=90)
plt.xlabel('Repos')
plt.title('Distribuição de Estrelas e Forks - Jogos')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
x = np.arange(len(web_names))

plt.bar(x - width/2, web_stars, width, label='Estrelas (Web)', color='red')
plt.bar(x + width/2, web_forks, width, label='Forks (Web)', color='pink')

plt.xticks(x, web_names, rotation=90)
plt.xlabel('Repos')
plt.title('Distribuição de Estrelas e Forks - Web')
plt.legend()
plt.show()
```
